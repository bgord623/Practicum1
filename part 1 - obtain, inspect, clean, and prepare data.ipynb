## Cleaning & preparing data
- For consistency and to trim the datasets, the timeframe for all dataframes will be set to 1985-2016
- Stats for each playerID will be combined

```python
# preparing batting data
df = df[(df.yearID >= 1985) & (df.yearID <= 2016)].reset_index(drop=True)
```

```python
# combining stats
df = df.groupby(['playerID'])
df = df.sum().reset_index()
```

```python
df.head()
```

- The stats were summed correctly but 'yearID' and 'stint' must be dropped
- Note that only numerical columns remain after grouping by playerID

```python
drop = ['yearID', 'stint']
df = df.drop(columns=drop, axis=1)
```

```python
df.info()
```

```python
# preparing fielding data
fielding_df = fielding_df[(fielding_df.yearID >= 1985) & (fielding_df.yearID <= 2016)].reset_index(drop=True)
```

```python
fielding_df.head()
```

- Per the documentation, the last 5 columns are specific to the catcher position and will be filtered out

```python
fielding_df = fielding_df.loc[:, :'DP']
fielding_df.head()
```

- If the fielding stats were combined by playerID now the positions would be lost
- Before combining them, the most frequently played position of their career will be extracted into a new dataframe for EDA and machine learning at later stages

```python
# extracting positions from fielding_df based on appearances at said positions
position_df = pd.DataFrame()

fielder_id = fielding_df.playerID.value_counts().index

# loop through fielding_df playerIDs and group positions 
for i in fielder_id:
    temp_df1 = fielding_df[fielding_df['playerID'].isin([i])]
    fielder_pos = temp_df1['POS'].value_counts().index
    temp_df2 = pd.DataFrame()
    
    # loop through positions and identify most frequently occurring position per player
    for v in fielder_pos:
        temp_df1 = temp_df1[temp_df1['POS'].isin([v])]
        temp_df2 = temp_df2.copy().append({'POS':v,'G':temp_df1.sum()['G']},ignore_index=True)
     
    # returns index (playerID) of highest game count   
    temp_pos = temp_df2['G'].idxmax()
    position_df = position_df.append({'playerID':i,'POS':temp_df2['POS'][temp_pos]},ignore_index=True)
```

```python
position_df.head()
```

```python
# combining stats
fielding_df = fielding_df.groupby(['playerID'])
fielding_df = fielding_df.sum().reset_index()

# drop useless columns; also dropping 'G' since it is included with batting stats
drop = ['yearID', 'stint', 'G']
fielding_df = fielding_df.drop(columns=drop, axis=1)
```

- Fielding data is ready for merging

```python
# preparing salary data
salaries.head()
```

```python
# average salary per year (2007-2016)
avg_sal = salaries.groupby(['yearID']).mean()
avg_sal.tail(10)
```

- To account for this inflation, salaries will be adjusted based on the change in combined league salary per year, relative to combined 2016 league salary

```python
# combined league salary per year
sal_yr = salaries.groupby(['yearID']).sum().reset_index()
sal_yr.head()
```

```python
# determine change per year
sal_yr['multiplier'] = ''

for i in range(len(sal_yr)):
    
    numerator = sal_yr.iloc[-1][1]
    denominator = sal_yr.iloc[i][1]
    
    sal_yr['multiplier'][i] = numerator / denominator
```

```python
sal_yr.head(10)
```

```python
def Adjust_Salary(full_df, multiplier_df):
    '''
    Adjusts salaries based on total change per year relative to 2016 total;
    Function uses default column names from seah lahman datasets
    '''    
    df1 = full_df
    df2 = multiplier_df

    df1['adj_salary'] = ''
    years = df1.yearID.value_counts().index


    for i in range(len(df1)):
        for year in years:
            if df1.yearID[i] == year:
                mult = df2[df2.yearID == year].multiplier
                exist_sal = df1.salary[i]
                df1['adj_salary'][i] = exist_sal * mult.values[0]
```

```python
adj_sal = salaries.copy()

Adjust_Salary(adj_sal, sal_yr) 
```

```python
adj_sal.adj_salary = adj_sal.adj_salary.astype('int64')
```

```python
# combining salaries
salaries_df = adj_sal.groupby(['playerID'])
salaries_df = salaries_df.sum().reset_index()

# drop useless columns
drop = ['yearID']
salaries_df = salaries_df.drop(columns=drop, axis=1)
```

- There are 7 records with no defensive stats

```python
# updating missing positions
no_pos = df_all[df_all.POS.isna()]
no_pos.POS = ['OF', 'OF', 'OF', 'OF','OF', '1B','1B']

df_all[df_all.POS.isna()] = no_pos
```

```python
# Missing values reflect players who had no defensive statistics and no values present; to be replaced with zero
df_all = df_all.replace(np.nan, 0)
```

```python
df_all.info()
```

- Great, no more missing values. For visual purposes the full names of players will be added and columns repositioned

```python
# adding player names to dataframe 
name_df = names[['playerID', 'nameFirst', 'nameLast']]
name_df['name'] = name_df['nameFirst'] + ' ' + name_df['nameLast']
name_df = name_df[['playerID', 'name']]
```

```python
name_df.head()
```

```python
df_all = pd.merge(df_all, name_df, on=['playerID'], how='left')
```

```python
df_all.head()
```

```python
def move_cols(df):
    
    '''
    update "columns" & "drops" lists before calling this function
    df will be reconfigured to show "columns" first and drop "drops"
    '''
    
    columns = ['name', 'salary', 'adj_salary', 'POS']   
    drops = ['playerID']
    
    df = df.copy()
    temp_df = df[columns].copy()
    df = df.drop(columns, axis=1)
    df = df.drop(drops, axis=1)
    new_df = pd.concat([temp_df, df], axis=1)
    
    return new_df
```

```python
df_final = move_cols(df_all)
```
